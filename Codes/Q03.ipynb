```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the images
arch_img = cv2.imread('005.jpg')
flag_img = cv2.imread('flag.png')

# Make a copy of the original architectural image for blending
arch_img_copy = arch_img.copy()

# Resize flag image if needed
flag_height, flag_width = flag_img.shape[:2]

# Create a list to store the points
points_architecture = []
points_flag = np.array([[0, 0], [flag_width - 1, 0], [flag_width - 1, flag_height - 1], [0, flag_height - 1]], dtype=np.float32)

# Mouse callback function to get points
def get_points(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        points_architecture.append((x, y))
        # Draw the points on the image just for visual feedback
        cv2.circle(arch_img, (x, y), 5, (0, 255, 0), -1)
        cv2.imshow("Architecture Image", arch_img)

# Display the architectural image and collect points
cv2.imshow("Architecture Image", arch_img)
cv2.setMouseCallback("Architecture Image", get_points)

# Wait until 4 points are selected
while len(points_architecture) < 4:
    cv2.waitKey(1)

# Convert selected points to NumPy array
points_architecture = np.array(points_architecture, dtype=np.float32)

# Compute the homography matrix
H, _ = cv2.findHomography(points_flag, points_architecture)

# Warp the flag image to the architectural image
warped_flag = cv2.warpPerspective(flag_img, H, (arch_img_copy.shape[1], arch_img_copy.shape[0]))

# Blend the flag onto the original architectural image (without circles)
blended = cv2.addWeighted(arch_img_copy, 1, warped_flag, 0.5, 0)

# Close the OpenCV window
cv2.destroyAllWindows()

# Convert BGR to RGB for display in Jupyter Notebook
blended_rgb = cv2.cvtColor(blended, cv2.COLOR_BGR2RGB)


fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Plot the first image
ax[0].imshow(cv2.cvtColor(arch_img, cv2.COLOR_BGR2RGB))
ax[0].set_title("Image 1")
ax[0].axis("off")

# Plot the second image
ax[1].imshow(cv2.cvtColor(flag_img, cv2.COLOR_BGR2RGB))
ax[1].set_title("Image 2")
ax[1].axis("off")

# Display the figure with both images
plt.tight_layout()
plt.show()

# Plot the blended image in the notebook
plt.imshow(blended_rgb)
plt.title("Blended Image")
plt.axis('off')  # Hide axes
plt.show()
```
